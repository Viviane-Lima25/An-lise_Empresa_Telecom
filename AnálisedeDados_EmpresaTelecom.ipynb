```python
plt.figure(figsize=(10, 6))
sns.histplot(dados['tenure'], bins=30, kde=True, color='skyblue')
plt.title('Distribuição do Tempo de Permanência dos Clientes (tenure)', fontsize=16)
plt.xlabel('Tempo de Permanência (meses)')
plt.ylabel('Número de Clientes')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Boxplot para clientes que NÃO deram churn
axs[0].boxplot(dados.loc[dados['Churn'] == 'No', 'Charges.Monthly'])
axs[0].set_title('Cobranças Mensais - Clientes que NÃO deram Churn')
axs[0].set_xlabel('Churn = No')
axs[0].set_ylabel('Cobrança Mensal (Charges.Monthly)')
axs[0].grid(True)

# Boxplot para clientes que deram churn
axs[1].boxplot(dados.loc[dados['Churn'] == 'Yes', 'Charges.Monthly'])
axs[1].set_title('Cobranças Mensais - Clientes que deram Churn')
axs[1].set_xlabel('Churn = Yes')
axs[1].set_ylabel('Cobrança Mensal (Charges.Monthly)')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tenure', y='Charges.Monthly', data=dados, hue='Churn', alpha=0.7, palette='coolwarm')
plt.title('Relação entre Tempo de Permanência e Cobranças Mensais', fontsize=16)
plt.xlabel('Tempo de Permanência (meses)')
plt.ylabel('Cobrança Mensal (Charges.Monthly)')
plt.legend(title='Churn')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Boxplot das Cobranças Mensais por Churn (separado)
axs[0].boxplot([
    dados.loc[dados['Churn'] == 'No', 'Charges.Monthly'],
    dados.loc[dados['Churn'] == 'Yes', 'Charges.Monthly']
])
axs[0].set_title('Cobranças Mensais por Churn')
axs[0].set_xlabel('Churn')
axs[0].set_ylabel('Cobrança Mensal')
axs[0].set_xticklabels(['No', 'Yes'])
axs[0].grid(True)

# Boxplot do Tempo de Permanência por Churn (separado)
axs[1].boxplot([
    dados.loc[dados['Churn'] == 'No', 'tenure'],
    dados.loc[dados['Churn'] == 'Yes', 'tenure']
])
axs[1].set_title('Tempo de Permanência por Churn')
axs[1].set_xlabel('Churn')
axs[1].set_ylabel('Tempo de Permanência (meses)')
axs[1].set_xticklabels(['No', 'Yes'])
axs[1].grid(True)

plt.tight_layout()
plt.show()

```

#Principais Resultados

** Cobranças Mensais e Tempo de Permanência**

Clientes que deram churn tendem a ter cobranças mensais ligeiramente mais altas do que clientes que permaneceram.

O tempo de permanência é significativamente menor para clientes que cancelaram, indicando que muitos abandonam o serviço antes de completar 1 ano.

**Distribuição do Tempo de Permanência**
Há uma concentração de cancelamentos no primeiro mês de contrato.

Outro pico de concentração está próximo ao prazo final de 72 meses, possivelmente associado ao encerramento de contratos de longo prazo.

**Cobranças Anuais por Churn**
O boxplot das Cobranças Anuais revelou outliers expressivos, principalmente entre clientes que cancelaram.

Esses valores atípicos indicam cobranças anuais muito acima do padrão observado, chegando a mais de R$ 8.000,00.

**Possíveis Causas dos Outliers**
A presença de valores tão elevados de cobrança anual pode estar relacionada a:

Erros de registro ou digitação – valores incorretos lançados no sistema.

Cobrança retroativa acumulada – clientes que atrasaram pagamentos e tiveram faturas acumuladas.
